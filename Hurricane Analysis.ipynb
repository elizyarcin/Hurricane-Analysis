4. You believe that knowing how often each of the areas of the Atlantic are affected by these strong hurricanes is important for making preparations for future hurricanes.

   Write a function that counts how often each area is listed as an affected area of a hurricane. Store and return the results in a dictionary where the keys are the affected areas and the values are counts of how many times the areas were affected.
   
   Test your function on your hurricane dictionary.

````python
# 4
# Counting Damaged Areas
def count_damaged_areas(hurricane_dict):
    area_counts = {}

    for hurricane in hurricane_dict.values():
        areas_affected = hurricane['Areas Affected']

        for area in areas_affected:
            if  area not in area_counts:
                area_counts[area] = 1
            else :
                area_counts[area] += 1
    return area_counts
# create dictionary of areas to store the number of hurricanes involved in
pprint(count_damaged_areas(hurricanes))
````

6. Write a function that finds the hurricane that caused the greatest number of deaths, and how many deaths it caused.

````python
# 6
# Calculating the Deadliest Hurricane

def deadliest_hurricane(hurricane_dict):
    max_deaths = -1  # Başlangıç için çok küçük bir sayı
    deadliest = None

    for hurricane in hurricane_dict.values():
        deaths = hurricane['Deaths for each Hurricane']
        if deaths > max_deaths:
            max_deaths = deaths
            deadliest = hurricane['Name']
    
    return deadliest, max_deaths

# find highest mortality hurricane and the number of deaths
result = deadliest_hurricane(hurricanes)
print(f"The most deathlist hurricane is : {result[0]}, death number: {result[1]}")
````

7. Just as hurricanes are rated by their windspeed, you want to try rating hurricanes based on other metrics.

   Write a function that rates hurricanes on a mortality scale according to the following ratings, where the key is the rating and the value is the upper bound of deaths for that rating.
   
   ```py
   mortality_scale = {0: 0,
   1: 100,
   2: 500,
   3: 1000,
   4: 10000}
   ```
   
   For example, a hurricane with a `1` mortality rating would have resulted in greater than `0` but less than or equal to `100` deaths. A hurricane with a `5` mortality would have resulted in greater than `10000` deaths.
   
   Store the hurricanes in a new dictionary where the keys are the mortaility ratings and the values are lists containing a dictionary for each hurricane that falls into that mortality rating.
   
   Test your function on your hurricane dictionary.

````python
# 7
# Rating Hurricanes by Mortality
def rate_hurricanes_by_mortality(hurricane_dict, mortality_scale):
    rated_hurricanes = {key: [] for key in mortality_scale.keys()}
    rated_hurricanes[max(mortality_scale.keys()) + 1] = []
    
    for hurricane in hurricane_dict.values():
        deaths = hurricane['Deaths for each Hurricane']
        
        # scale for death number
        rating = None
        for scale, upper_bound in sorted(mortality_scale.items()):
            if deaths <= upper_bound:
                rating = scale
                break
        # no scale more then 1000 , put biggest  scale 
        if rating is None:
            rating = max(mortality_scale.keys()) + 1
        
        rated_hurricanes[rating].append(hurricane)
    
    return rated_hurricanes

# categorize hurricanes in new dictionary with mortality severity as key
mortality_scale = {0: 0, 1: 100, 2: 500, 3: 1000, 4: 10000}

rated = rate_hurricanes_by_mortality(hurricanes, mortality_scale)

from pprint import pprint
pprint(rated)
````

````python
# 8
# Calculating Hurricane Maximum Damage

def find_highest_damage(hurricane_dict):
    max_damage = -1
    max_hurricane = None

    for hurricane in hurricane_dict.values():
        damage = hurricane['Damages  of hurricane']
      
        if isinstance(damage, (int, float)):
            if damage > max_damage:
                max_damage = damage
                max_hurricane = hurricane['Name']

    return max_hurricane, max_damage



# find highest damage inducing hurricane and its total cost
hurricane_name, damage_amount = find_highest_damage(hurricanes)
print(f"The highest damage hurricane : {hurricane_name}, its total cost: ${damage_amount:,.2f}")
````

9. Lastly, you want to rate hurricanes according to how much damage they cause.

   Write a function that rates hurricanes on a damage scale according to the following ratings, where the key is the rating and the value is the upper bound of damage for that rating.
   ```py
   damage_scale = {0: 0,
   1: 100000000,
   2: 1000000000,
   3: 10000000000,
   4: 50000000000}
   ```
   
   For example, a hurricane with a `1` damage rating would have resulted in damages greater than `0` USD but less than or equal to `100000000` USD. A hurricane with a `5` damage rating would have resulted in damages greater than `50000000000` USD (talk about a lot of money).
   
   Store the hurricanes in a new dictionary where the keys are damage ratings and the values are lists containing a dictionary for each hurricane that falls into that damage rating.
   
   Test your function on your hurricane dictionary.

````python
# 9
# Rating Hurricanes by Damage
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}

# categorize hurricanes in new dictionary with damage severity as key
def rate_hurricanes_by_damage(hurricane_dict):
    damage_scale = {
        0: 0,
        1: 100_000_000,
        2: 1_000_000_000,
        3: 10_000_000_000,
        4: 50_000_000_000
    }

    # Sonuçları tutacak sözlük, rating: [hurricane_dict,...]
    rated_hurricanes = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}  # 5. seviye için ayrıca oluşturduk.

    for hurricane in hurricane_dict.values():
        damage = hurricane['Damages  of hurricane']

        # Hasar bilgisi sayısal mı kontrol et (string olan "Damages not recorded" olabilir)
        if not isinstance(damage, (int, float)):
            continue  # Kayıt yoksa atla

        # Damage değerine göre rating belirle
        rating = 5  # Başlangıçta en yüksek rating (50 milyar üstü)
        for rate, upper_bound in damage_scale.items():
            if damage <= upper_bound:
                rating = rate
                break

        # Fırtınayı ilgili rating listesine ekle
        rated_hurricanes[rating].append(hurricane)

    return rated_hurricanes

# Fonksiyonu test et
from pprint import pprint
pprint(rate_hurricanes_by_damage(hurricanes))
````
